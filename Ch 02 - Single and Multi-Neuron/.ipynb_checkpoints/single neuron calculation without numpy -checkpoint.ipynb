```python
ip =[1,2,3]
wt =[0.2,0.8,-0.5]
bias = 2
```

# Output calculation 

```python
op= (ip[0]*wt[0] +
     ip[1]*wt[1] +
     ip[2]*wt[2])+bias
```

```python
print(op)
```
