```python
# Neural network from Scratch
```

```python
import numpy as np

# Inputs
X = np.array([1.0, 2.0])

# Weights
W = np.array([0.5, -1.5])

# Bias
b = 0.2

# Linear combination
z = np.dot(X, W) + b

# ReLU activation
output = np.maximum(0, z)

print("Neuron output:", output)
```

# Multiple Neurons

```python
X = np.array([1.0, 2.0])

W = np.array([
    [0.5, -1.5],
    [1.0, 0.5],
    [-0.3, 0.8]
])

b = np.array([0.2, -0.1, 0.5])

z = np.dot(W, X) + b
output = np.maximum(0, z)

print("Layer output:", output)
```
